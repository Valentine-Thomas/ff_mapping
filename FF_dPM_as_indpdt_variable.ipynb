# Expressing variables in function dpm

```python
#files containing fusion events folder
path= '/mnt/c/Users/Valentine/Documents/Analysis/POI_replicats/' 
files = ['Prm1','Dni1','Dni2','Fus1','Exo70', 'Myo51', 'Cdc8','Exo84'] 
Myo= ['Myo52_M', 'Myo52_P']
RefM= '_B'
RefP= '_R'
```

```python
from trajalign.traj import Traj
from trajalign.average import load_directory
import numpy as np
from matplotlib import pyplot as plt
import os
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import norm as norm
```

```python
def replicats_nb(POI):
    rep_list = [ folder for folder in os.listdir(path+POI+'/aligned_trajectories/') if 'Rep_' in folder ]
    r= len(rep_list)
    return r

def get_data(POI,rep) :
    directory_rep = os.listdir( path+POI+'/aligned_trajectories/Rep_'+str(rep+1)+'/') #absolute path of directory to charge data from
    f = [ folder for folder in directory_rep if 'cell_' in folder ] # 'cell_' -> folder nam to list
    return f 

n= len(files) 
Rep= []
for i in range(n):
    r=replicats_nb(files[i])
    Rep.append(r)



```

```python
def dist_as_indpd_v (POI_d_P_M, POI_var,POI,rep, what='fluo'):
    list=[]
    cell=0
    print(POI)
    for i in range (len(POI_d_P_M)):
        
        dist_traj= POI_d_P_M[i]
        var_traj= POI_var [i]
        inter=set(dist_traj.t()).intersection(var_traj.t())  # get list of common frames between the two trajs
        dist=[] 
        var=[]
        time=[]
        ct=[]    
        for i in inter:
            i= int(i)
            ct.append(i)
        ct.sort()
       
        for f in ct:
            if -60<f<20:
                for x in range (len(dist_traj)):
                
                    if dist_traj.t(x)==f:
                        time.append(f)
                        dist.append(-(dist_traj.f(x)))
                     
                for x in range (len(var_traj)):
                    if var_traj.t(x)==f:
                        var.append(var_traj.f(x))
        if len(time)>0:    
            T=Traj()
            T.input_values('frames', time)
            T.input_values('t', dist)
            T.input_values('f', var)
            T.save(path+POI+'/aligned_trajectories/Rep_'+str(rep+1)+'/cell_%03d/fdPM_%s.txt'%(cell, what))
            list.append(T)
        cell+=1
        
    
    return list
```

# distance P_POI
#### load data

```python
folders=[[] for _ in range(n)] #lists for POI rep folders => Subsets: POI then Rep
Cells=[[] for _ in range(n)] #lists of trajectories for each events for each POI of each replicate

for i in range(len(Rep)):
    folders[i]=[[] for _ in range(Rep[i])] #prep placeholder for directory of each replicats
    Cells[i]=[[] for _ in range(Rep[i])] 

# list all cell_ folders for each POI 

for i in range(n) :
    for r in range(Rep[i]):
        f=get_data(files[i],r)
        folders[i][r]= f 

ct=0
for f in folders:
    r=1
    i=0 
    for rep in f:
        for event in rep: 
            t_l = load_directory(path+files[ct] +'/aligned_trajectories/Rep_'+str(r)+'/'+ event ,pattern= 'dir_dist', comment_char = '#', t_unit='min' , frames = 0 ,t= 1, f = 2)
            Cells[ct][i].extend(t_l)
        r+=1
        i+=1
    ct+=1
    
    
#Split lists per distance measured 
distP_M=[[] for _ in range(n)]
distP_POI=[[]for _ in range(n)]
distM_POI=[[]for _ in range(n)]
for i in range(len(Rep)):
    distP_M[i]=[[] for _ in range(Rep[i])] 
    distP_POI[i]=[[] for _ in range(Rep[i])] 
    distM_POI[i]=[[] for _ in range(Rep[i])] 
    
for i in range(n):
        for r in range (len(Cells[i])):
            for t in Cells[i][r]:
                if 'P_M' in t.annotations() ['file'] : 
                    distP_M[i][r].append(t)
                elif 'P_POI' in t.annotations() ['file'] :
                    distP_POI[i][r].append(t)
                else : 
                    distM_POI[i][r].append(t)
```

```python
ddist_PPOI=[[] for _ in range(n)]
for i in range(len(Rep)):
    ddist_PPOI[i]=[[] for _ in range(Rep[i])] 
for i in range (len(distP_M)):
    for r in range(len(distP_M[i])):
        ddist_PPOI[i][r]=dist_as_indpd_v(distP_M[i][r], distP_POI[i][r],files[i],r, what='distP_POI')
```
