```python
#files containing fusion events folder
path= '/mnt/c/Users/Valentine/Documents/Analysis/POI_replicats/' 
files = ['Prm1','Dni1','Dni2','Fus1','Exo70', 'Myo51', 'Cdc8','Exo84'] 

from trajalign.traj import Traj
from trajalign.average import load_directory
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import os
from scipy.stats import norm as norm

def replicats_nb(POI):
    rep_list = [ folder for folder in os.listdir(path+POI+'/aligned_trajectories/') if 'Rep_' in folder ]
    r= len(rep_list)
    return r

def get_data(POI,rep) :
    directory_rep = os.listdir( path+POI+'/aligned_trajectories/Rep_'+str(rep+1)+'/') #absolute path of directory to charge data from
    f = [ folder for folder in directory_rep if 'cell_' in folder ] # 'cell_' -> folder nam to list
    return f 

n= len(files) 
Rep= []
for i in range(n):
    r=replicats_nb(files[i])
    Rep.append(r)
```

## Loading data
load all fdPM files in one => slipt them in lists according to what/color

```python
folders=[[] for _ in range(n)] #lists for POI rep folders => Subsets: POI then Rep
Cells=[[] for _ in range(n)] #lists of trajectories for each events for each POI of each replicate

for i in range(len(Rep)):
    folders[i]=[[] for _ in range(Rep[i])] #prep placeholder for directory of each replicats
    Cells[i]=[[] for _ in range(Rep[i])] 

# list all cell_ folders for each POI 

for i in range(n) :
    for r in range(Rep[i]):
        f=get_data(files[i],r)
        folders[i][r]= f 

ct=0
for f in folders:
    r=1
    i=0 
    for rep in f:
        for event in rep: 
            t_l = load_directory(path+files[ct] +'/aligned_trajectories/Rep_'+str(r)+'/'+ event ,pattern= 'fdPM', comment_char = '#', t_unit='min' , frames = 0 ,t= 1, f = 2)
            Cells[ct][i].extend(t_l)
        r+=1
        i+=1
    ct+=1

    
#split in fonction file name
f_P=[[] for _ in range(n)]
f_M=[[] for _ in range(n)]
f_POI=[[] for _ in range(n)]
Ecc_P=[[] for _ in range(n)]
Ecc_M=[[] for _ in range(n)]
Ecc_POI=[[] for _ in range(n)]
dist_POI=[[] for _ in range(n)]

for i in range (len(Rep)):
    f_P[i]=[[] for _ in range(Rep[i])]
    f_M[i]=[[] for _ in range(Rep[i])]
    f_POI[i]=[[] for _ in range(Rep[i])]
    Ecc_P[i]=[[] for _ in range(Rep[i])]
    Ecc_M[i]=[[] for _ in range(Rep[i])]
    Ecc_POI[i]=[[] for _ in range(Rep[i])]
    dist_POI[i]=[[] for _ in range(Rep[i])]
    
for i in range(n):
        for r in range (len(Cells[i])):
            for t in Cells[i][r]:
                if 'distP' in t.annotations() ['file'] : 
                    dist_POI[i][r].append(t)
                elif 'ecc' in t.annotations() ['file'] :
                    if '_P' in t.annotations() ['file'] :
                        Ecc_P[i][r].append(t)
                    elif '_M' in t.annotations() ['file']:
                        Ecc_P[i][r].append(t)
                    else:
                        Ecc_POI[i][r].append(t)
                elif 'fluo' : 
                    if '_P' in t.annotations() ['file'] :
                        f_P[i][r].append(t)
                    elif '_M' in t.annotations() ['file']:
                        f_M[i][r].append(t)
                    else:
                        f_POI[i][r].append(t)
```

```python
def mad (s):
    list_s=[]
    m=np.median(s)
    for x in s: 
        s_i=np.absolute(x-m)
        list_s.append(s_i)
   
    mad_s= np.median(list_s)
    mad=mad_s*1.4826
    return mad


def av_dist_var(list_var, names, nR,what='f', sample='rep'):
    print(names)
    if sample=='tot':
        list_var=[item for sublist in list_var for item in sublist]
    colnames=['dPM','x_plt','Median Var','Standard Deviation', 'Variance']
    matrice={}
    for k in range (len (colnames)):
        matrice[colnames[k]] = []   
    hist=[[] for _ in range(7)]
    index = 0
    binvect = [-800,-700,-600,-500,-400,-300,-200,50]
    for ev in range(7):
        var=[]
        bin_min= binvect[index]
        bin_max=binvect[index+1]-1
        for traj in list_var :
            for ct in range (len(traj)):
                if bin_min<traj.t(ct)<bin_max:
                    var.append(traj.f(ct))     
        matrice['dPM'].append('[%d-%d]'%(bin_min,bin_max)) 
        matrice['x_plt'].append((bin_min+bin_max)/2)
        var=np.array(var)
        var = var[~np.isnan(var)] 
        
        if (var.size>0):         
            matrice['Median Var'].append(np.median(var))
            matrice['Standard Deviation'].append(mad(var))
            matrice['Variance'].append(mad(var)**2)
            hist[index]=var
        else: 
            matrice['Median Var'].append(float('Nan'))     
            matrice['Standard Deviation'].append(float('Nan'))     
            matrice['Variance'].append(float('Nan'))     
        index+=1  
        ##cannot save using Traj function => to transform into panda df and save this way
    E=pd.DataFrame(matrice)
    E=E.set_index('dPM') 
    if sample=='rep':
        E.to_csv(path+names+'/aligned_trajectories/Data/average/Average_d_%s_Rep_%s.csv'%(what,str(nR+1)))
    if sample=='tot':
        E.to_csv(path+names+'/aligned_trajectories/Data/average/Average_d_%s.csv'%what)
    return E,hist
```

```python
def plot_hist (av, hist,name,rep,what ='x',type='fluo'):     
    os.makedirs(path+name+'/aligned_trajectories/Data/average/hist/', exist_ok=True)
    if (what=='x'):
        hist= hist[0]
        time=av.t()
        med= av.coord()[0]
        err= av.coord_err()[0]
        
        
    if (what=='y'):
        hist= hist[1]
        time=av.t()
        med= av.coord()[1]
        err= av.coord_err()[1]
        
    if (what== 'f') :
        hist= hist[2]
        med= av.f()
        err= av.f_err()
        time=av.t()      
          
    if (what== 'dir') :
        med= av['Median Var'].tolist()
        err= av['Standard Deviation'].tolist()
        time=av['x_plt'].tolist()
    n = 0
    j = 0
    k = 27
    length= int(np.ceil(len(hist)/27))
    
    for l in range (length):    
        h= hist[j:k]
        m= med[j:k]
        e=err [j:k]
        t=time[j:k]
        plt.figure(figsize=(13, 4*9))
        for i in range (len(h)):
            plt.subplot(9,3,i+1) 
            plt.hist(h[i], bins=10)
            plt.grid() 
            a=m[i]+e[i]
            b=m[i]-e[i]                    
            plt.axvline(m[i], color = 'red')
            plt.axvline(a, color = 'hotpink', ls= '--')
            plt.axvline(b, color = 'hotpink', ls ='--')
            plt.title(t[i]) 
        figname=path+name+'/aligned_trajectories/Data/average/hist/%s_%s_hist_dPM_%s_%d.pdf'%(name,str(rep+1),type,n)
        plt.savefig(figname)
        plt.close()
        j+=27
        n+=1
        k+= 27 
```

```python
av_dist= [[] for _ in range(n)]
hist_dist=[[]for _ in range(n)]
for k in range(len(Rep)):
    av_dist[k]= [[] for _ in range(Rep[k])]
    hist_dist[k]=[[]for _ in range(Rep[k])]

for i in range(len(dist_POI)):
    for r in range (len(dist_POI[i])):
        av_dist[i][r], hist_dist[i][r]=av_dist_var(dist_POI[i][r], files[i],r,what='P_POI',sample='rep')
        plot_hist(av_dist[i][r],hist_dist[i][r],files[i],r, what='dir',type='dist')
```
